```python
def auc_roc(y_true, y_pred):
    # can be any tensorflow metric
    value, update_op = tensorflow.contrib.metrics.streaming_auc(y_pred, y_true)
    # find all variables created for this metric
    metric_vars = [i for i in tensorflow.local_variables() if 'auc_roc' in i.name.split('/')[1]]
    
    # Add metric variables to GLOBAL_VARIABLES collection.
    # They will be initialized for new session.
    for v in metric_vars:
        tensorflow.add_to_collection(tensorflow.GraphKeys.GLOBAL_VARIABLES, v)
    
    # force to update metric values
    with tensorflow.control_dependencies([update_op]):
        value = tensorflow.identity(value)
    return value
```

```python
def create_model():
    #learn_rate=0.01
    #momentum=0.001
    dropout_rate=0.3
    model = Sequential()
    model.add(Dense(512, activation='relu',kernel_regularizer=regularizers.l2(0.01), input_shape=(inputs_shape,)))
    model.add(Dropout(dropout_rate))
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(128,activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))
    #optimizer=Adam(lr=learn_rate, momentum=momentum)
    model.compile(loss='binary_crossentropy', optimizer='Adam',metrics=['accuracy',auc_roc])
    #model.compile(loss='binary_crossentropy', optimizer='Adam',metrics=['accuracy',RocCallback])
    return model
```

```python
# Wrap Keras model so it can be used by scikit-learn
kears_estimator = KerasClassifier(build_fn=create_model, verbose=1)
```

```python
#Parameters for GridSearching
param_grid = {
        'model__batch_size': [10, 20, 40, 60, 80, 100],
        'model__epochs':[10, 50, 100]
}

# Estimator Pipeline
pipeline = Pipeline(
      [
        ('std_scaler', StandardScaler()),
        ('model',    kears_estimator),
      ]
    )

# Cross Validation Splitter
#cv_group_splitter = GroupKFold(n_splits=5)

cv_group_splitter=5


#Perform GridSearch
search = GridSearchCV(
    estimator=pipeline, 
    param_grid=param_grid,
    scoring = ['roc_auc', 'accuracy'],
    refit = 'roc_auc', 
    cv=cv_group_splitter,
    verbose = 1
    )


grid_result = search.fit(df_features, mind_wandered_label,groups=participant_ids)

```

```python
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
