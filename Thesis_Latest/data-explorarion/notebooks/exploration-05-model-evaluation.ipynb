```python
# globally useful imports of standard libraries needed in this notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# metric functions from scikit-learn
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve

# import project specific modules used in this notebook
import sys
sys.path.append('../src')
import mindwandering.data
```

# Model Evaluation

The reference paper mainly used Area Under the Receiver Operating Characteristic (AUROC) scores to evaluate their model performance.  These
are also known as Area Under the Curve of the Receiver Operating Characteristics (AUC-ROC).  ROC is related to concepts like precision and recall.
The ROC is the relation of false positive rate (FPR) to true positive rate (TPR) as the threshold for deciding true/false (for a binary
classification) is varied.  The following is a good quick summary 

[Understnding AOC-ROC Curve](https://towardsdatascience.com/understanding-auc-roc-curve-68b2303cc9c5)

In this notebook we will develop the tools we need to evaluate the models being built to replicate the reference paper.

We will be evaluating random and other models using the correct labels, so load the correct labels.

```python
# mindwandered label
df_label = mindwandering.data.get_df_label()
df_label = df_label.mind_wandered_label
```

# Random Model

A model that guesses at random will have no real discrimination capacity.  When AUC is 0.5 this means the model is performing no better than
random guessing.

For our model evaluation we will be using leave-one-participant-out cross validation.  This mean we may need to be gathering our own predictions
to then evaluate results.

Lets simulate a random model, and evaluate its accuracy, confusion matrix, precision score, recall score (also known as true positive rate TPR),
specificity, and false positive rate, and the auc score.  We will also plot precision vs. recall plots, and the AUC-ROC curve.

```python
# create random predictions
num_trials = df_label.shape[0]

# random predictions
predictions = np.random.random_sample(num_trials) > 0.5

# for precision and recall curves we need the prediction probabilities.  Use 0.5 for completely random predictions
true_random_probabilities = np.ones(num_trials) * 0.5

# for our random predictor, create probabilities above/below 0.5 a bit based on the prediction
probabilities = ((predictions.astype(int) / 5) - 0.1) * (np.random.random_sample(num_trials) / 5) + 0.5
```

```python
# display scores
accuracy = accuracy_score(df_label, predictions)
recall = recall_score(df_label, predictions)
precision = precision_score(df_label, predictions)
aucroc = roc_auc_score(df_label, predictions)

print('Accuracy score:        ', accuracy)
print('Recall (TPR) score:    ', recall)
print('Precision score:       ', precision)
print('AUC-ROC score:         ', aucroc)
```

```python
# display confusion matrix
cm = confusion_matrix(df_label, predictions)
print(cm)

# extract tp, tn, fp, fn
tp = cm[1,1]
tn = cm[0,0]
fp = cm[0,1]
fn = cm[1,0]

# true positive rate and false positive rate
tpr = tp / (tp + fn)
fpr = fp / (tn + fp)

# specificity
specificity = tn / (tn + fp)

print("True positives:  ", tp)
print("True negatives:  ", tn)
print("False positives: ", fp)
print("False negatives: ", fn)
print("True Positive Rate:  ", tpr)
print("False Positive Rate: ", fpr)
print("Specificity:         ", specificity)
# want to plot the confusion matrix, but we cant use the trained predictor, need to use
# the gathered predictions
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot();
```

```python
# plot precision/recall curve
precision, recall, thresholds = precision_recall_curve(df_label, probabilities)

plt.figure(figsize=(12,8))
plt.plot(recall[1:], precision[1:], label='Model precision vs. recall')
plt.legend()
plt.xlabel('Recall (Positive label: 1)')
plt.ylabel('Precision (Positive label: 1)');
```

```python
# plot ROC AUC curves
true_random_fpr, true_random_tpr, _ = roc_curve(df_label, true_random_probabilities)
random_fpr, random_tpr, _ = roc_curve(df_label, probabilities)

plt.figure(figsize=(12,8))
plt.plot(true_random_fpr, true_random_tpr, label='True Random performance')
plt.plot(random_fpr, random_tpr, label='Random Model performance')
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate');
```

Lets make a function that will calculate and display all of this evaluation information, including
the precision/recall and AUC-ROC figures.  It will be useful to have this and not repeat 
whenever we want to evaluate the performance metrics.

```python
def evaluate_model_results(df_label, predictions, probabilities):
    """Given the correct labels, and a set of predictions, evaluate the model.  We
    also need the classifier probabilities, the probability that it is the
    positive (true or 1) class.
    
    We calculate and display accuracy, precision, recall, AUC-ROC score, etc.
    We also plot the precision/recall curve and the AUC-ROC curve.
    """
    # display scores
    accuracy = accuracy_score(df_label, predictions)
    recall = recall_score(df_label, predictions, zero_division=0)
    precision = precision_score(df_label, predictions, zero_division=0)
    aucroc = roc_auc_score(df_label, predictions)

    print('Accuracy score:        ', accuracy)
    print('Recall (TPR) score:    ', recall)
    print('Precision score:       ', precision)
    print('AUC-ROC score:         ', aucroc)
    print('')
    
    # display confusion matrix properties
    cm = confusion_matrix(df_label, predictions)

    # extract tp, tn, fp, fn
    tp = cm[1,1]
    tn = cm[0,0]
    fp = cm[0,1]
    fn = cm[1,0]

    # true positive rate and false positive rate
    tpr = tp / (tp + fn)
    fpr = fp / (tn + fp)

    # specificity
    specificity = tn / (tn + fp)

    print('True positives:  ', tp)
    print('True negatives:  ', tn)
    print('False positives: ', fp)
    print('False negatives: ', fn)
    print('True Positive Rate (recall): ', tpr)
    print('False Positive Rate:         ', fpr)
    print('Specificity:                 ', specificity)
    print('')
    
    print(cm)
    
    # plot the confusiion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot();    
    
    # plot the precision/recall curve
    precision, recall, thresholds = precision_recall_curve(df_label, probabilities)

    plt.figure(figsize=(12,8))
    plt.plot(recall[1:], precision[1:], label='Model precision vs. recall')
    plt.legend()
    plt.xlabel('Recall (Positive label: 1)')
    plt.ylabel('Precision (Positive label: 1)');
    
    # plot the AUC-ROC curve
    num_trials = df_label.shape[0]
    random_probabilities = np.zeros(num_trials)
    random_fpr, random_tpr, _ = roc_curve(df_label, random_probabilities)
    model_fpr, model_tpr, _ = roc_curve(df_label, probabilities)

    plt.figure(figsize=(12,8))
    plt.plot(random_fpr, random_tpr, label='Random performance')
    plt.plot(model_fpr, model_tpr, label='Model performance')
    plt.legend()
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate');    
```

```python
evaluate_model_results(df_label, predictions, probabilities)
```

# Imbalanced Model Predictors

We can do better than random here since the model is imbalanced (about 78% of trials are labeled false).
If you just predict false, you get the following AUC-ROC curve and scores.

Notice accuracy is 72%, but recall is 0 because never get any of the true labels correct.

Also note that the auc-roc is still 0.5, just like a random guesser.  So while accuracy can be misleading,
auc-roc above 0.5 for imbalanced classes can be helpful in identifying if model is doing anything
or not.

```python
predictions = np.zeros(num_trials)
probabilities = np.zeros(num_trials) # probability is alwyas 0 for the positive class.

evaluate_model_results(df_label, predictions, probabilities)
```

Likewise, if you always predict 1, then you get perfect recall, because you recall all of the
true positives.  But accuracy is only 27% and AUC-ROC scores is also still 0.5.

```python
predictions = np.ones(num_trials)
probabilities = np.ones(num_trials) # probability is alwyas 0 for the positive class.

evaluate_model_results(df_label, predictions, probabilities)
```

# Perfect Predictors

A perfect predictor will get 100% accuracy, which will lead to no false positive or false negatives.

```python
predictions = df_label.values
probabilities = df_label.values # probability is alwyas 0 for the positive class.

evaluate_model_results(df_label, predictions, probabilities)
```

Likewise, a predictor that is always wrong will get all false positive and false negatives.

```python
predictions = np.abs(df_label.values - 1)
probabilities = np.abs(df_label.values - 1)

evaluate_model_results(df_label, predictions, probabilities)
```
